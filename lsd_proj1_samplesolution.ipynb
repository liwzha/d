## Problem 2: Compute Gists & Problem 3: Calculate Pairwise Distances

Starting from here, we will use a few image sets created by the students.

```python
### Read in the images
import string 

pairs = []
with open('./lsdproj1_images/pairs.txt','r') as f:
    pairs = map(lambda line: line.strip().split(','), f.readlines())
    f.close()

# choose a few datasets
pairs = [['true_motorcycles','true_trees'],
['true_dolphins','true_tomatoes'],
['true_bunnies','true_hamburgers'],
['true_guitars','true_penguins'],
['true_buddhas','true_elephants'],
['true_plant','true_water'],
['true_bicycle','true_car'],
['true_pumpkins','true_brain']]

imagesets = []
# For each pair of images, read in the indices of images
for p in pairs:
    noun = ['','']
    true_ids = [[],[]]
    ids = [[],[]]
    
    # Read in the indicies of labeled images
    for i in range(2):
        with open('./lsdproj1_images/%s.txt'%p[i],'r') as true_f:
            true_ids[i] = [int(ind) for ind in true_f.readlines()[0].strip('[]\n').split(',')]
            true_f.close()
        
        noun[i] = p[i].split('_')[1].translate(string.maketrans('',''),string.digits)
        with open ('./lsdproj1_images/%s.txt'%noun[i],'r') as f:
            ids[i] = [int(ind) for ind in f.readlines()[0].strip('[]\n').split(',')]
            f.close()
            
    imagesets.append({'noun':noun,'true_ids':true_ids,'ids':ids})
```

```python
# functions for computing gist vectors and pair-wise distance
import numpy as np
import scipy
import pickle
from scipy import spatial

def gist(index):
    import DAL
    import scipy
    import leargist
    
    tinyimages = DAL.create('tinyimages')
    img = scipy.misc.toimage( \
    tinyimages.byid(index).reshape(32,32,3, order="F").copy())
    return leargist.color_gist(img)

def standardize(vec):
    import numpy as np
    return (vec - np.mean(vec)) / np.std(vec)   
```

```python
''' 
This cell computes pair-wise distances between images in each data set and then pickle the results for later use.
The computation will take a while.
Information of all image sets will be stored in the list 'imagesets'.
'''
for set_ind in xrange(len(imagesets)):
    ids_pair = imagesets[set_ind]['ids']
    ids = ids_pair[0] + ids_pair[1]
    gists = dview.map_sync(gist,ids)
    gists_np = np.array(gists)
    gists_np = dview.map_sync(standardize, gists_np)
    
    _dist_matrix = scipy.spatial.distance.pdist(gists_np,'euclidean')
    dist_matrix = scipy.spatial.distance.squareform(_dist_matrix)
    imagesets[set_ind]['dist_matrix'] = dist_matrix
    
    # pickle the data for later use
    pickle.dump(imagesets[set_ind],open('./pickle/lsdproj1-imagesets-%s-%s'%(imagesets[set_ind]['noun'][0],imagesets[set_ind]['noun'][1]),'wb'))

imageset_noun = [ imageset['noun'] for imageset in imagesets]
imagesets = [] # for reducing memory usage
```

```python
'''
Show a few examples of "distance" in gist space.
'''

import random
import numpy

# randomly choose n images from each of the two categories in an image set 
# and show their k nearest neighbors and k farthest neighbors in gist space
def show_knn( imageset, k, n ):
    import string
    import DAL
    tinyimages = DAL.create('tinyimages')
    
    print '===============================[ %s and %s ]==============================='%(imageset['noun'][0], imageset['noun'][1])
    for t in xrange(2):
        ids = imageset['ids'][t]
        samples_locids = [ ids.index(r) for r in random.sample(imageset['true_ids'][t],n)]
        
        dist_matrix = imageset['dist_matrix']
        for i in samples_locids:
            sub_dist_matrix = dist_matrix[t*2000:(t+1)*2000,i]
            sortedarg = sub_dist_matrix.argsort()
            knn_locids = sortedarg[:(k+1)]
            kfn_locids = sortedarg[-k:]
            image_ids = [ids[j] for j in knn_locids]
            images = tinyimages.byid(image_ids)
            print 'template:'
            tinyimages.display([images[0]])
            print 'nearest neighbours in gist space:'
            tinyimages.display(images[1:])
            print 'farthest neighbours in gist space:'
            images = tinyimages.byid([ids[j] for j in kfn_locids])
            tinyimages.display(images)

k,n = 15,1
print '%d nearest neighbours and %d farthest neighbours in gist space: '%(k,k)
for noun in imageset_noun:
    imageset = pickle.load(open('./pickle/lsdproj1-imagesets-%s-%s'%(noun[0],noun[1]),'rb'))
    show_knn( imageset, k, n )
```

The outputs above show some samples images and their nearest/farthest neighbors in the gist space.  From each image set that contains tiny images of two categories, one image is sampled from each of the categories, and it is used as a template.  For each template, we search for its 15 nearest neighbors and 15 farthest neighbors in the gist space.  As we can see from the results, the images that are considered closer to the template in the gist space are very likely to share the same structures.  Images of "guitars" and "buddhas" are two examples.  We can expect that gist representation performs better in matching up images that are perceptually similar than raw pixel representation does, as raw pixel representation is highly sensitive to translation and rotation -- a translation or rotation of objects in the image could change the values of quite a few pixels and result in a large difference in raw pixel space.

## Problem 4: Build a k-NN classifier

```python
'''
Define functions for running knn clasifier and making the plots.
'''

# this function randomly select a few images from the data set as training and test data
def sample_data(imageset, l, testSize=100):
    import random
    allids = imageset['ids'][0] + imageset['ids'][1]
    true_ids = imageset['true_ids']
    sample_true_ids = [ random.sample(ids, (l+100)/2 ) for ids in true_ids ]
    train_ids = [ idx[:(l/2)] for idx in sample_true_ids ]
    train_ids = [ allids.index(i) for i in train_ids[0]+train_ids[1] ]
    test_ids = [ idx[(l/2):((l/2)+testSize) ] for idx in sample_true_ids ]
    test_ids = [ allids.index(i) for i in test_ids[0]+test_ids[1] ]
    non_train_ids = [ j for j in range(len(allids)) if not j in train_ids ]
    return allids, train_ids, test_ids, non_train_ids
    
def knn_test(imageset, k, l, num_tests=50):
    import random
    import numpy as np
    
    # knn classifier
    def knn_predict(labeled_images, test_image, dist_matrix):
        labels = np.array([0.0]*(l/2) + [1.0]*(l/2))
        knn_ind = np.argsort( dist_matrix[labeled_images, test_image] )[:k]
        predict = sum( labels[knn_ind] ) > (k/2)
        return int(predict)

    def one_test(true_ids, ids, dist_matrix):
        
        # randomly choose training and test data
        labeled_ids,test_ids = [[],[]], [[],[]]
        allids, labeled_locids, test_locids, dummy = sample_data(imageset, l) 
        
        # classify the test data
        predictions = np.array(map( lambda x: knn_predict(labeled_locids, x, dist_matrix), test_locids ))
        
        # compute error rate
        true_test_labels = np.zeros(100)
        true_test_labels[50:] = 1
        error_rate = sum(predictions != true_test_labels) / 100.0
        
        return error_rate
        
    error_rates = []
    true_ids = imageset['true_ids']
    ids = imageset['ids']
    dist_matrix = imageset['dist_matrix']
    for t in range(num_tests):
        error_rates.append(one_test(true_ids, ids, dist_matrix))
    return (np.mean(error_rates), np.std(error_rates))

# function for making plots.  called in "show_results".
def show_errbars( err_std_pairs, krange, k_name, lrange, nouns ):
    import matplotlib.pyplot as plt
    N = len(err_std_pairs)
    fig, axs = plt.subplots(nrows=1, ncols=N, figsize=(3*len(krange), 2))
    for i in xrange(N):
        k = krange[i]
        ave_err = map(lambda x: x[0], err_std_pairs[i])
        std_err = map(lambda x: x[1], err_std_pairs[i])
        ax = axs[i]
        ax.errorbar(lrange, ave_err, yerr=std_err, fmt='o-')
        ax.set_xlim(lrange[0]-5, lrange[-1]+5)
        ax.set_title('%s=%s'%(k_name, krange[i]))
        
    fig.text(0.5, 1.0,'%s and %s'%(nouns[0],nouns[1]), fontsize = 18,horizontalalignment='center')    
    plt.tight_layout()
    

def show_results(knn_exp):
    import matplotlib.pyplot as plt

    krange = knn_exp['krange']
    lrange = knn_exp['lrange']
    
    for key,val in knn_exp['res'].iteritems():
        nouns = key
        show_errbars( val, krange, knn_exp['k_name'], lrange, nouns )
```

## Problem 5: Construct a kNN graph & Problem 6: Compute the SSL harmonic function

```python
'''
Define functions for building a kNN graph
'''

# this function compute the connected components of a graph based on its adjacency matrix.
def get_graph_components(adjmatrix):
    import numpy as np
    
    # use breadth-first search to find the component containing the given nodes
    def bfs(adjmatrix, nodes):
        comp = set()
        while len(nodes)>0:
            v = nodes.pop()
            if v not in comp:
                comp.add(v)
                nodes += list(np.nonzero(adjmatrix[v,:])[0])
        return comp

    N = len(adjmatrix)
    unchecked_nodes = set(range(N))
    components_list = []
    while len(unchecked_nodes) > 0 :
        node = unchecked_nodes.pop()
        component = bfs(adjmatrix, [node])
        components_list.append(component)
        unchecked_nodes -= component
    return components_list

# connect the rest vertices of a graph with their nearest neighour in the component
def connect_graph_components(dist_matrix, adjmatrix, component):
    import numpy as np
    N = len(dist_matrix)
    nodes_in_comp = list(component)
    nodes_out_comp = list(set(range(N)) - set(nodes_in_comp))
    for j in nodes_out_comp:
        nn_ind = np.argsort(dist_matrix[nodes_in_comp,j])[0]
        nn_ind = nodes_in_comp[nn_ind]
        adjmatrix[nn_ind, j] = adjmatrix[j, nn_ind] = 1.0
    adjmatrix -= np.diag(np.diag(adjmatrix))
    return adjmatrix


# Find the minimum k that give a kNN graph where at least {percent}% of the nodes are connected into a single component.
# Connect the remaining nodes to their nearest neighbor in the connected component.
# 
# input: 
# dist_matrix: a matrix of pair-wise distance
# percent: the objective size of largest component in percentage
#
# output:
# k: the minimun k
# adjmatrix: adjacency matrix of the connected kNN graph
def compute_knn_graph(dist_matrix, percent=0.95):
    import numpy as np
    N = len(dist_matrix)
    adjmatrix = np.zeros(dist_matrix.shape)
    sortedind = np.argsort(dist_matrix,axis=0)
    k = 0
    while(1):
        k += 1
        for i in xrange(N):
            knn_ind = sortedind[:(k+1), i]
            adjmatrix[i,knn_ind] = adjmatrix[knn_ind,i] = 1
        
        components_list = get_graph_components(adjmatrix)
        max_comp_ind = np.argmax( map( lambda x:len(x), components_list ) )
        if len(components_list[max_comp_ind]) >= percent*N:
            adjmatrix = connect_graph_components(dist_matrix, adjmatrix, components_list[max_comp_ind])
            break
            
    return k, adjmatrix

def get_knn_adjmatrix(dataset):
    dist_matrix = imageset['dist_matrix']
    k, adjmatrix = compute_knn_graph(dist_matrix)
    print 'optimal k for dataset %s and %s: %d '%(imageset['noun'][0],imageset['noun'][1],k)
    return (k, adjmatrix)
```

```python
'''
Define functions for testing SSL classifier that users harmonic functions
'''

def ssl_classifier( adjmatrix, l_ind, u_ind, f_l, style='median' ):
    import numpy as np
    W = np.array(adjmatrix,dtype=float)
    D = np.diag(W.sum(axis=1))
    Duu = D[np.ix_(u_ind,u_ind)]
    Wuu = W[np.ix_(u_ind,u_ind)]
    Wul = W[np.ix_(u_ind,l_ind)]
    f_u = np.linalg.solve(Duu-Wuu, Wul*f_l)
    
    if style == 'median':
        threshold = np.median(f_u,axis=0)[0]
    else:
        threshold = 0.5
    f_u = [ int(t>threshold) for t in f_u ]
    return f_u
    
def ssl_tester( imageset, adjmatrix, l, style='median', num_test=10 ):
    import random
    import numpy as np
    u = 4000 - l
    ids = imageset['ids'][0] + imageset['ids'][1]
    error_rate = []
    for t in range(num_test):
        
        allids, labeled_locind, test_locind, unlabeled_locind = sample_data(imageset, l)
        
        f_l = np.matrix( [0.0]*(l/2) + [1.0]*(l/2) ).reshape((l,1))
        f_u = ssl_classifier( adjmatrix, labeled_locind, unlabeled_locind, f_l, style )
        
        error = 0
        for j,v in enumerate(unlabeled_locind):
            if v in test_locind:
                true_label = int( v >= 2000 )
                if f_u[j] != true_label:
                    error += 1
        
        error_rate.append( error/100.0 )
    return (np.mean(error_rate), np.std(error_rate))

```

```python
'''
Run test SSL classifier on a few image sets.
'''

krange = ['0.5', 'median']
lrange = range(10,60,10)
ssl_exp = {'krange':krange, 'k_name':'threshold', 'lrange':lrange, 'res':dict()}
for noun in imageset_noun[:5]:
    imageset = pickle.load(open('./pickle/lsdproj1-imagesets-%s-%s'%(noun[0],noun[1]),'rb'))
    optk, adjmatrix = get_knn_adjmatrix(imageset)
    ssl_exp['res'][(imageset['noun'][0],imageset['noun'][1])] = []
    for k in krange:
        oneset = []
        for l in lrange:
            ave_std_err = ssl_tester( imageset, adjmatrix, l, k, 10 )
            oneset.append(ave_std_err)
        ssl_exp['res'][(imageset['noun'][0],imageset['noun'][1])].append(oneset)
    
show_results(ssl_exp)
```

The plots above show the result of implementing semi-supervised learning using harmonic functions on 5 datasets.  Two different values of thresholds are used: 0.5 and median (the median of all estimated function values).  Using median as threshold is to take advantage of the knowledge that there are 50% images of each category. For each l, the number of labeled data, the experiment is repeated for 10 times, each time we randomly select a subset of true images as labeled data and another subset for testing.  The average and std of these error rates are plotted as error bars in figures above.  By comparing the error rates, we can see that using median as the threshold leads to an improvement in accuracy.  Meanwhile, in all plots, the error rates tends to be reducing as the number of labeled data grows.  Yet, based on the results we have seen so far, the error rates of SSL classifier (using median as threshold) is approximately on the same level as those produced by k-NN classifier.
